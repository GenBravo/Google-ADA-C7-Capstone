### Modeling

Add as many cells as you need to conduct the modeling process.

```python
### YOUR CODE HERE ###
df2 = pd.get_dummies(df1, drop_first=True)
df2.head()
```

```python
df2['left'].value_counts()
```

```python
# Separate the dataset into labels (y) and features (X).

y = df2['left']
X = df2.drop(columns='left')
```

```python
# Separate into train, validate, test sets.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=0)

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.25, stratify = y_train, random_state = 0)
```

```python
print(X_tr.shape, X_val.shape, X_test.shape,'\n', y_tr.shape, y_val.shape, y_test.shape)
```

> ### *Model 1: Random Forest*

```python
rf = RandomForestClassifier(random_state=23)
```

```python
rf_params = {'n_estimators' : [25, 50, 100, 150, 200, 300],
            'max_depth' : [5, 10, 15, 20, 25, 40, 50],
            'min_samples_split' : [0.001, 0.005, 0.01],
            'min_samples_leaf' : [0.5,1],
            'max_features' : ['sqrt'],
            'max_samples' : [.5,.9]}
```

```python
#Create list of split indices.

split_index = [0 if x in X_val.index else -1 for x in X_train.index]
custom_split = PredefinedSplit(split_index)
```

```python
rf_val = GridSearchCV(estimator=rf,
                      param_grid=rf_params,
                      verbose = 1,
                      n_jobs=-1,
                      refit='f1',
                      cv=custom_split)
```

```python
%%time
# Fit the model.

rf_val.fit(X_train, y_train)
```

```python
rf_val.best_params_
```

```python
# New RF classifier with optimal params found by GridSearchCV.

rf_opt = RandomForestClassifier(max_depth = 15,
                                 max_features = 'sqrt',
                                 max_samples = 0.9,
                                 min_samples_leaf = 1,
                                 min_samples_split = 0.005,
                                 n_estimators = 100)
```

```python
# Fit the optimal model.

rf_opt.fit(X_train, y_train)
```

```python
# Predict on test set.

rf_y_pred = rf_opt.predict(X_test)
```

```python
# Scores

rf_pc_test = precision_score(y_test, rf_y_pred)
print("The precision score is {pc:.3f}".format(pc = rf_pc_test))

rf_rc_test = recall_score(y_test, rf_y_pred)
print("The recall score is {rc:.3f}".format(rc = rf_rc_test))

rf_ac_test = accuracy_score(y_test, rf_y_pred)
print("The accuracy score is {ac:.3f}".format(ac = rf_ac_test))

rf_f1_test = f1_score(y_test, rf_y_pred)
print("The F1 score is {f1:.3f}".format(f1 = rf_f1_test))
```

```python
#pickle the model

with open('my_rf_cv_model.pickle', 'wb') as to_write:
    pickle.dump(rf_val, to_write)
```

```python
rf_val.best_score_
```

> ### *Model 2: XGBoost*

```python
xgb = XGBClassifier(objective='binary:logistic', random_state = 23)
```

```python
cv_params = {'max_depth': [3, 4, 5, 6],
            'min_child_weight': [3, 5],
            'learning_rate': [0.1, 0.2, 0.3],
            'n_estimators': [5,10,15],
            'subsample': [0.6, 0.7, 0.8, 0.9],
            'colsample_bytree': [0.6, 0.7, 0.8, 0.9]
            }
```

```python
scoring = {'accuracy', 'precision', 'recall', 'f1'}
```

```python
xgb_cv = GridSearchCV(xgb,
                      cv_params,
                      scoring = scoring,
                      cv = 5,
                      refit = 'f1')
```

```python
%%time
# fit the GridSearch model to training data

xgb_cv = xgb_cv.fit(X_train, y_train)
```

```python
xgb_cv
```

```python
#pickle the model

with open('my_xgb_cv_model.pickle', 'wb') as to_write:
    pickle.dump(xgb_cv, to_write)
```

```python
xgb_cv.best_params_
```

```python
# New XGB classifier with optimal params found by GridSearchCV.

xgb_opt = XGBClassifier(colsample_bytree = 0.7,
                         learning_rate = 0.3,
                         max_depth = 6,
                         min_child_weight = 3,
                         n_estimators = 15,
                         subsample = 0.9)
```

```python
# Fit the optimal model.

xgb_opt.fit(X_train, y_train)
```

```python
# Predict on test set.

xgb_y_pred = xgb_opt.predict(X_test)
```

```python
# Scores

xgb_pc_test = precision_score(y_test, xgb_y_pred)
print("The precision score is {pc:.3f}".format(pc = xgb_pc_test))

xgb_rc_test = recall_score(y_test, xgb_y_pred)
print("The recall score is {rc:.3f}".format(rc = xgb_rc_test))

xgb_ac_test = accuracy_score(y_test, xgb_y_pred)
print("The accuracy score is {ac:.3f}".format(ac = xgb_ac_test))

xgb_f1_test = f1_score(y_test, xgb_y_pred)
print("The F1 score is {f1:.3f}".format(f1 = xgb_f1_test))
```

## Recall evaluation metrics

- **AUC** is the area under the ROC curve; it's also considered the probability that the model ranks a random positive example more highly than a random negative example.
- **Precision** measures the proportion of data points predicted as True that are actually True, in other words, the proportion of positive predictions that are true positives.
- **Recall** measures the proportion of data points that are predicted as True, out of all the data points that are actually True. In other words, it measures the proportion of positives that are correctly classified.
- **Accuracy** measures the proportion of data points that are correctly classified.
- **F1-score** is an aggregation of precision and recall.
